```python
daily_df = trans_df.drop(["item_description","invoice_id","customer_id"],axis = 1).dropna()
daily_df["income"] = daily_df["item_price"]*daily_df["quantity_sold"]
daily_df = daily_df.drop("item_price",axis = 1)
daily_df
```

```python
dates = daily_df.transaction_date.unique()
dates.sort()
weeks = daily_df.transaction_week.unique()
weeks.sort()
```

```python
store_df = pd.read_csv("/kaggle/input/data-storm-4/Store-info.csv")
store_df_weekly = store_df.copy() 
store_df_weekly
```

```python
for date in dates:
    day_df = daily_df.loc[daily_df["transaction_date"]==date]
    store_df[str(date)]=''
    store_df[str(date)+"S"]=''
    for shop in store_df["shop_id"]:
        shop_df = day_df.loc[day_df["shop_id"]==shop]
        day_income = shop_df.income.sum()
        day_sales = shop_df.quantity_sold	.sum()
        store_df[str(date)][store_df.index[store_df["shop_id"]==shop]] = day_income
        store_df[str(date)+"S"][store_df.index[store_df["shop_id"]==shop]] = day_sales
store_df
```

```python
for week in weeks:
    week_df = daily_df.loc[daily_df["transaction_week"]==week]
    store_df_weekly[str(week)]=0
    store_df_weekly[str(week)+"S"]=0
    for shop in store_df["shop_id"]:
        shop_df = week_df.loc[week_df["shop_id"]==shop]
        week_income = shop_df.income.sum()
        week_sales = shop_df.quantity_sold.sum()
        store_df_weekly[str(week)][store_df_weekly.index[store_df_weekly["shop_id"]==shop]] = week_income
        store_df_weekly[str(week)+"S"][store_df_weekly.index[store_df_weekly["shop_id"]==shop]] = week_sales
store_df_weekly
```

```python
store_df_weekly.info()
```

```python
store_df_weekly["income"]=0
store_df_weekly["sales"]=0
for shop in store_df["shop_id"]:
    shop_df = daily_df.loc[daily_df["shop_id"]==shop]
    income = shop_df.income.sum()
    sales = shop_df.quantity_sold.sum()
    store_df_weekly["income"][store_df_weekly.index[store_df_weekly["shop_id"]==shop]] = income
    store_df_weekly["sales"][store_df_weekly.index[store_df_weekly["shop_id"]==shop]] = sales
store_df_weekly
```

```python
#X = store_df_weekly.drop(["shop_profile","shop_id"],axis = 1).head(100)
X = store_df_weekly.drop(["shop_profile","shop_id"],axis = 1).head(100)
y = store_df_weekly["shop_profile"].head(100)
#to_be_tested = store_df_weekly.drop(["shop_profile","shop_id"],axis = 1).tail(24)
to_be_tested = store_df_weekly.drop(["shop_profile","shop_id"],axis = 1).tail(24)
X
```
